```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('mymoviedb.csv' , lineterminator = '\n')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Genre'].head()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtypes)
```

```python
df['Release_Date'] = df['Release_Date'].dt.year
df['Release_Date'].dtypes
```

```python
df.head()
```

```python
df['Vote_Average'] = df['Vote_Average'].apply(lambda x: 'Popular' if x > 7.5 else ('Average' if x > 5 else ('Below_avg' if x > 2.5 else 'Not_popular')))
```

```python
df.head()
```

```python
df['Vote_Average'].value_counts()
```

```python
df["Genre"] = df["Genre"].astype(str).apply(lambda x: x.split(", "))
df = df.explode("Genre").reset_index(drop=True)

df.head()
```

```python
df.info()
```

```python
df.nunique()
```

# DATA VISUALIZATION

```python
sns.set_style('whitegrid')
```

#  Q1.What is the most frequent genre of movies released on Netflix 

```python
df['Genre'].describe()
```

```python
sns.catplot(y='Genre', data = df, kind = 'count',
            order = df['Genre'].value_counts().index,
            color = '#4287f5')
plt.title('Genre Column Distribution')
plt.show()
```

# Q2. Which is the highest votes in vote_average column 

```python
sns.catplot(y='Vote_Average', data = df, kind = 'count',
            order = df['Vote_Average'].value_counts().index,
            color = '#FFA500')

plt.title('Title Board Distribution')
plt.show()
            
```

# Q3. What movie got the highest popularity and what its genre ?

```python
df[df['Popularity'] == df['Popularity'].max()]
```

# Q4. What movie got the lowest popularity and what is its genre ?

```python
df[df['Popularity'] == df['Popularity'].min()]
```

# Q5. Which year has the most filmmed movies ?

```python
df['Release_Date'].hist()
plt.title("Release Date Column Distribution")
plt.show()
```
